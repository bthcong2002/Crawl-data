```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
page_source = BeautifulSoup(driver.page_source)
# print(page_source)
```

```python
job_names = page_source.find_all(name = 'h3', attrs = {'class' : 'title job-details-link-wrapper'})
list_names = []
list_links = []
for name in job_names:
    job = name.find(name = 'a')
    list_names.append(job.text.strip())
    list_links.append('https://itviec.com/'+ job['href'])
```

```python
list_names[:]
```

```python
list_links[:]
```

```python
# Lấy tên Công ty tuyển dụng:
list_company = []
for i in list_links:
    # print(i)
    page_response = requests.get(i)
    page_content = BeautifulSoup(markup=page_response.content, features="html.parser")

    company_names = page_content.find_all(name = 'h3', attrs = {'class' : 'employer-long-overview__name hidden-xs d-none d-sm-block'})

    for name in company_names:
        job = name.find('a')
        list_company.append(job.text.strip())
```

```python
list_company[:]
```

```python
list_skills = []
for i in list_links:
    a = []
    page_response = requests.get(i)
    page_content = BeautifulSoup(markup=page_response.content, features="html.parser")

    skill_require = page_content.find_all(name = 'a', attrs = {'class' : 'big ilabel mkt-track'})

    for skill in skill_require:
        a.append(skill.text.replace('\n', ''))

    list_skills.append(a)
```

```python
list_skills[:]
```

```python
list_salary = []
import re

salary = page_source.find_all(name = 'div', attrs = {'class' : 'job__description'})

for a in salary:
    b = a.find(name = 'div', attrs = {'class' : 'svg-icon__text'}).text.strip()
    c = re.sub('USD|m|,','',b)
    d = re.sub("Thỏa thuận|You'll love it","0 - 0",c)
    e = re.split(' - ', d)
    list_salary.append(e)
```

```python
list_salary[:]
```

```python
data_document = pd.DataFrame({
    'Name' : list_names,
    'Company' : list_company,
    'Link' : list_links,
    'Skills require' : list_skills,
    'Salary': list_salary,
})
```
